```python
import numpy as np

A = np.random.randint(1, 10, size=(3, 3))
B = np.random.randint(1, 10, size=(3, 3))

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
```

```python
addition = A + B
subtraction = A - B
multiplication = A @ B

print("A + B:\n", addition)
```

```python
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)

print("Determinant of A:", det_A)
print("Rank of A:", rank_A)
if det_A != 0:
    inverse_A = np.linalg.inv(A)
    print("\nInverse of A:\n", inverse_A)
else:
    print("\nMatrix A is singular and has no inverse.")
```

```python
import matplotlib.pyplot as plt
vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

print("Vector 1:\n", vector1)
print("\nVector 2:\n", vector2)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(vector1, marker='o', label='Vector 1')
plt.plot(vector2, marker='s', label='Vector 2')

plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Two Random Vectors")
plt.legend()
plt.grid(True)

plt.show()
```
